```python
import pandas as pd
import numpy as np

import statsmodels.graphics.api as smg
from statsmodels.graphics.tsaplots import plot_acf
```

生成する乱数の数を設定

```python
n = 1000
```

qを設定

```python
q = 3
```

$\epsilon$を生成

```python
eps = np.random.normal(0, 1, n)
```

```python
X = []
for i in range(n - q):
  X.append(np.mean(eps[i: i + q]))
```

```python
plot_acf(X, lags=29)
```
